```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
h = 0.7
omega_l = 0.7
omega_r = 4.3 * h**(-2) * 1e-5
omega_m= 0.3 - omega_r 
```

```python
z = np.logspace(-1, 4, 1000)
```

```python
om_r = omega_r * (1 + z)**4
om_m = omega_m * (1 + z)**3
om_l = omega_l + z*0
total = om_r + om_m + om_l
```

```python
plt.plot(om_r,om_m)
```

```python
plt.figure()
plt.xlabel('Redshift')
plt.ylabel('Omega(z)')
plt.plot(z,om_l, 'b', label = r'\Omega_\Lambda(z)')
plt.plot(z,om_r, 'k', label = r'\Omega_R(z)')
plt.plot(z,om_m, 'g', label = r'\Omega_{NR}(z)')
plt.plot(z,total, 'r', lw=2, label = 'Total')
plt.legend(loc='upper left')
plt.show()
```
